3. Prove (AB)

T = BTAT

```python
A = np.array([[3, 4],
              [7, 8]])

B = np.array([[5, 3],
              [2, 1]])
AB = np.dot(A, B)
AB_T = AB.T
print("(AB)^T =")
print(AB_T)
```

```python
B_T = B.T
A_T = A.T
BT_AT = np.dot(B_T, A_T)
print("B^T * A^T =")
print(BT_AT)
```

Solve the following system of Linear equation using Inverse Methods.

2x − 3y + z = −1
x − y + 2z = −3
3x + y − z = 9

{Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}

```python
A = np.array([[2, -3, 1],
              [1, -1, 2],
              [3, 1, -1]])

B = np.array([-1, -3, 9])
A_inv = np.linalg.inv(A)
X = np.dot(A_inv, B)
print("Solution [x, y, z] =")
print(X)
```

4.2 Experiment: How Fast is Numpy?
In this exercise, you will compare the performance and implementation of operations using plain Python
lists (arrays) and NumPy arrays. Follow the instructions:
1. Element-wise Addition:
• Using Python Lists, perform element-wise addition of two lists of size 1, 000, 000. Measure
and Print the time taken for this operation.
• Using Numpy Arrays, Repeat the calculation and measure and print the time taken for this
operation.

```python
import time

# Using Python Lists
size = 1000000
list1 = list(range(size))
list2 = list(range(size))

start_time = time.time()

# Element-wise addition using list comprehension
list_sum = [list1[i] + list2[i] for i in range(size)]

end_time = time.time()
print("Time taken using Python lists: {:.5f} seconds".format(end_time - start_time))

# Using NumPy Arrays
arr1 = np.arange(size)
arr2 = np.arange(size)

start_time = time.time()

# Element-wise addition using NumPy
arr_sum = arr1 + arr2

end_time = time.time()
print("Time taken using NumPy arrays: {:.5f} seconds".format(end_time - start_time))
```

Element-wise Multiplication

• Using Python Lists, perform element-wise multiplication of two lists of size 1, 000, 000. Mea-
sure and Print the time taken for this operation.

• Using Numpy Arrays, Repeat the calculation and measure and print the time taken for this
operation.

```python
import time

# Using Python Lists
size = 1000000
list1 = list(range(size))
list2 = list(range(size))

start_time = time.time()

# Element-wise multiplication using list comprehension
list_mul = [list1[i] * list2[i] for i in range(size)]

end_time = time.time()
print("Time taken using Python lists: {:.5f} seconds".format(end_time - start_time))

# Using NumPy Arrays
arr1 = np.arange(size)
arr2 = np.arange(size)

start_time = time.time()

# Element-wise multiplication using NumPy
arr_mul = arr1 * arr2

end_time = time.time()
print("Time taken using NumPy arrays: {:.5f} seconds".format(end_time - start_time))
```

3. Dot Product
• Using Python Lists, compute the dot product of two lists of size 1, 000, 000. Measure and
Print the time taken for this operation.
• Using Numpy Arrays, Repeat the calculation and measure and print the time taken for this
operation.

```python
import time

size = 1000000

# Using Python Lists
list1 = list(range(size))
list2 = list(range(size))

start_time = time.time()

# Dot product using sum of element-wise multiplication
dot_list = sum(list1[i] * list2[i] for i in range(size))

end_time = time.time()
print("Dot product using Python lists:", dot_list)
print("Time taken using Python lists: {:.5f} seconds".format(end_time - start_time))

# Using NumPy Arrays
arr1 = np.arange(size)
arr2 = np.arange(size)

start_time = time.time()

# Dot product using NumPy
dot_numpy = np.dot(arr1, arr2)

end_time = time.time()
print("Dot product using NumPy arrays:", dot_numpy)
print("Time taken using NumPy arrays: {:.5f} seconds".format(end_time - start_time))
```

4. Matrix Multiplication
• Using Python lists, perform matrix multiplication of two matrices of size 1000x1000. Measure
and print the time taken for this operation.

• Using NumPy arrays, perform matrix multiplication of two matrices of size 1000x1000. Mea-
sure and print the time taken for this operation.

```python
import time
import random

size = 1000
# Using Python Lists
# Create two 1000x1000 matrices with random integers
matrix1 = [[random.randint(0, 10) for _ in range(size)] for _ in range(size)]
matrix2 = [[random.randint(0, 10) for _ in range(size)] for _ in range(size)]

start_time = time.time()

# Matrix multiplication using nested loops
result_list = [[0 for _ in range(size)] for _ in range(size)]
for i in range(size):
    for j in range(size):
        for k in range(size):
            result_list[i][j] += matrix1[i][k] * matrix2[k][j]

end_time = time.time()
print("Time taken using Python lists: {:.5f} seconds".format(end_time - start_time))

# Using NumPy Arrays
arr1 = np.random.randint(0, 11, (size, size))
arr2 = np.random.randint(0, 11, (size, size))

start_time = time.time()

# Matrix multiplication using NumPy
result_numpy = np.dot(arr1, arr2)

end_time = time.time()
print("Time taken using NumPy arrays: {:.5f} seconds".format(end_time - start_time))
```
