Which adult male passenger (age ≥ 18 and Sex is ’male’) paid the highest fare relative to their
class?
To answer the question perform following operations:

class i.e. Fare / Pclass.

```python
dataset["fare_per_class"] = dataset["Fare"] / dataset["Pclass"]

print(dataset.head())
```

2. Subset rows where the passenger is male (Sex is ”male”) and an adult (Age is greater than or equal

```python
adult_males = dataset[(dataset["Age"] >= 18) & (dataset["Sex"] == "male")]

print(adult_males)
```

```python
adult_males_srt = adult_males.sort_values(by="fare_per_class", ascending=False)
print(adult_males_srt)
```

Calculate the total Fare paid across all passengers in the Titanic dataset.

```python
total_fare = dataset["Fare"].sum()

print("Total Fare paid by all passengers:", total_fare)
```

2. Subset for passengers in first class (Pclass is 1) and calculate their total fare.

```python
first_class_passengers = dataset[dataset["Pclass"] == 1]
total_fare_first_class = first_class_passengers["Fare"].sum()
print("Total Fare paid by first-class passengers:", total_fare_first_class)
```

3. Do the same for second class (Pclass is 2) and third class (Pclass is 3).

```python
second_class_passengers = dataset[dataset["Pclass"] == 2]
total_fare_second_class = second_class_passengers["Fare"].sum()

third_class_passengers = dataset[dataset["Pclass"] == 3]
total_fare_third_class = third_class_passengers["Fare"].sum()

print("Total Fare paid by second-class passengers:", total_fare_second_class)
print("Total Fare paid by third-class passengers:", total_fare_third_class)
```

Combine the fare totals from first, second, and third classes into a list.

```python
total_fares_by_class = [total_fare_first_class, total_fare_second_class, total_fare_third_class]

print(total_fares_by_class)
```

Divide the totals for each class by the overall total fare to get the proportion of fare revenue by class.

```python
overall_total_fare = dataset["Fare"].sum()
proportion_by_class = [fare_total / overall_total_fare for fare_total in total_fares_by_class]
print(proportion_by_class)
```

What percent of the total number of passengers on the Titanic belonged to each age group
(e.g., child, adult, senior)?
To answer the question perform following operation:
1. Create a new column, age group, that categorizes passengers into "child" (age < 18), "adult" (age
18{64), and "senior" (age 65 and above).

```python
def categorize_age(age):
    if age < 18:
        return "child"
    elif age < 65:
        return "adult"
    else:
        return "senior"

dataset["age_group"] = dataset["Age"].apply(categorize_age)


print(dataset[["Name", "Age", "age_group"]].head())
```

Calculate the total number of passengers on the Titanic.

```python
total_passengers = dataset.shape[0]

print("Total number of passengers on the Titanic:", total_passengers)
```

Count the number of passengers in each age group.

```python
age_group_counts = dataset["age_group"].value_counts()

print(age_group_counts)
```

Divide the count of each age group by the total number of passengers to get the proportion of passengers
in each age group.

```python
total_passengers = len(dataset)

age_group_counts = dataset["age_group"].value_counts()

age_group_proportion = age_group_counts / total_passengers

print(age_group_proportion)
```

Display the proportion as a percentage.

```python
age_group_percentage = (dataset["age_group"].value_counts() / len(dataset)) * 100

print(age_group_percentage)
```
