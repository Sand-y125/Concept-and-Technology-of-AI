```python
def factorial(n):
    
    if n < 0:
        raise ValueError("Factorial is only defined for non-negative integers.")
    
    # Base case
    if n == 0:
        return 1
    # Recursive case
    else:
        return n * factorial(n - 1)

print(factorial(5)) 
```

```python
def sum_nested_list(nested_list):
    """
    Calculate the sum of all numbers in a nested list.
    
    Args:
        nested_list (list): A list that may contain integers or other nested lists.
        
    Returns:
        int: The total sum of all integers in the nested list.
        
    Example:
        >>> sum_nested_list([1, [2, [3, 4], 5], 6, [7, 8]])
        36
    """
    total = 0
    for element in nested_list:
        if isinstance(element, list):  # Check if element is a list
            total += sum_nested_list(element)  # Recursive call
        else:
            total += element  # Add the number to the total
    return total

nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
print("Sum of nested list:", sum_nested_list(nested_list))

```

```python
def generate_permutations(s):
    if len(s) == 1:
        return [s]
    
    result = []
    
    for i in range(len(s)):
        char = s[i]
        remaining = s[:i] + s[i+1:]
        for perm in generate_permutations(remaining):
            result.append(char + perm)
    
    return result
print(generate_permutations("abc"))
print(generate_permutations("aab"))


```

```python
def calculate_directory_size(directory):
    
    total_size = 0
    
    for key, value in directory.items():
        if isinstance(value, dict):
            total_size += calculate_directory_size(value)
        else:
            total_size += value
            
    return total_size
directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}

print("Total directory size:", calculate_directory_size(directory_structure))

    
```

```python
def min_coins(coins, amount):
   
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0  
    
    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)
    
    return dp[amount] if dp[amount] != float('inf') else -1
coins = [1, 2, 5]
amount = 11
print("Minimum coins needed:", min_coins(coins, amount))  

coins2 = [2]
amount2 = 3
print("Minimum coins needed:", min_coins(coins2, amount2))  

```

```python
def longest_common_subsequence(s1, s2):
    
    m = len(s1)
    n = len(s2)
    
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])
    
    return dp[m][n]
s1 = "abcde"
s2 = "ace"

print("Length of LCS:", longest_common_subsequence(s1, s2)) 

```

```python
def knapsack(weights, values, capacity):
    n = len(weights)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]
    
    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            if weights[i - 1] <= w:
                dp[i][w] = max(values[i - 1] + dp[i - 1][w - weights[i - 1]],
                               dp[i - 1][w])
            else:
                dp[i][w] = dp[i - 1][w]
    
    return dp[n][capacity]

weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7

print("Maximum value achievable:", knapsack(weights, values, capacity))  
```
